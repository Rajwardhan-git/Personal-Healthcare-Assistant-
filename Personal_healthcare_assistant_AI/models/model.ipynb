```python
import pandas as pd
```

```python
dataset = pd.read_csv('E:\Personal_healthcare_assistant_AI\Personal_healthcare_assistant_AI\dataset/Training.csv')
```

```python
x=dataset.drop(['prognosis'], axis=1, inplace=False)
x
```

```python
y=dataset['prognosis']
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
le.fit(y)
y=le.transform(y)
```

```python
inverse_transform=le.inverse_transform(y)
inverse_transform
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=20)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar=StandardScaler()
scalar.fit(x_train)
x_train_scaled=scalar.transform(x_train)
x_train_scaled
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
from sklearn.svm import SVC 
```

```python
support_vc = SVC(kernel='linear')
support_vc.fit(x_train_scaled,y_train)
```

```python
x_test_scaled=scalar.transform(x_test)
```

```python
y_predict= support_vc.predict(x_test_scaled)
y_predict
```

```python
accuracy = accuracy_score(y_test, y_predict)
accuracy
```

```python
confusion_matrix(y_test,y_predict)
```

```python
confusion_matrix(y_test,y_predict)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(C=0.01, solver='liblinear')
model.fit(x_train,y_train)
```

```python
accuracy1= accuracy_score(y_test, model.predict(x_test))
accuracy1
```

```python
y_prediction =model.predict(x_test)
y_prediction
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test, y_prediction)
cm
```
